```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme(style = "whitegrid", palette = "pastel") 

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/cleaned_data.csv')
```

## Age Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(data['Age'], kde=True, bins=30)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

## Annual Income Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(data['Annual_Income'], kde=True, bins=30)
plt.title("Distribution of Annual Income")
plt.xlabel("Annual Income")
plt.ylabel("Count")
plt.show()
```

The distribution exhibits a strong positive (right) skew, meaning the tail extends far to the right. The majority of individuals earn lower incomes, while fewer individuals earn very high incomes.

## Income Boxplot

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=data, x='Annual_Income')
plt.title("Boxplot of Annual Income")
plt.xlabel("Annual Income")
plt.show()
```

## Outstanding Debt Boxplot

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=data, x='Outstanding_Debt')
plt.title("Boxplot of Outstanding Debt")
plt.xlabel("Outstanding Debt")
plt.show()
```

## Occupation Count

```python
plt.figure(figsize=(12,5))
sns.countplot(data=data, x='Occupation', palette = 'inferno')
plt.title("Count of Occupation Categories")
plt.xticks(rotation=45)
plt.show()
```

## Credit Mix Count

```python
plt.figure(figsize=(8,5))
sns.countplot(data=data, x='Credit_Mix', palette = 'Blues')
plt.title("Distribution of Credit Mix Types")
plt.xlabel("Credit Mix")
plt.ylabel("Count")
plt.show()
```

## Payment Behaviour Count

```python
plt.figure(figsize=(13,5))
sns.countplot(data=data, x='Payment_Behaviour', palette = 'Blues')
plt.title("Distribution of Payment Behaviour Types")
plt.xticks(rotation=60)
plt.show()
```

## Average Monthly Balance per Occupation

```python
plt.figure(figsize=(12,6))
colors = plt.cm.viridis(np.linspace(0, 1, len(data['Occupation'].unique())))

data.groupby('Occupation')['Monthly_Balance'].mean().sort_values().plot(
    kind='bar',
    color=colors
)

plt.title("Average Monthly Balance by Occupation")
plt.ylabel("Average Monthly Balance")
plt.xlabel("Occupation")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Income vs Outstanding Debt

```python
plt.figure(figsize=(10,6))
plt.hexbin(data['Annual_Income'], data['Outstanding_Debt'], gridsize=40)
plt.colorbar(label='Count')
plt.xlabel("Annual Income")
plt.ylabel("Outstanding Debt")
plt.title("Hexbin Plot")
plt.show()
```

## Credit Utilization vs Number of Loans

```python
plt.figure(figsize=(10,6))
plt.hist2d(data['Credit_Utilization_Ratio'], data['Num_of_Loan'], bins=40)
plt.colorbar(label='Count')
plt.xlabel("Credit Utilization Ratio")
plt.ylabel("Number of Loans")
plt.title("2D Histogram: Credit Utilization vs Loans")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='Num_of_Loan', y='Credit_Utilization_Ratio', palette="Set2")
plt.title("Credit Utilization by Loan Count")
plt.show()
```

## Interest Rate vs Number of Delayed Payments

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Interest_Rate', y='Num_of_Delayed_Payment', hue='Age')
plt.title("Interest Rate vs Number of Delayed Payments")
plt.xlabel("Interest Rate")
plt.ylabel("Delayed Payments")
plt.show()
```
